```python
import itertools
import pylab as pl
```

```python
def bubbleSort(arr):
    pasos=0
    comparaciones=0
    intercambios=0
    for k in range(len(arr) - 1, 0, -1):
        pasos+=1
        for i in range(k):
            if arr[i] > arr[i + 1]:
                comparaciones+=1
                temp = arr[i]
                arr[i] = arr[i + 1]
                arr[i + 1] = temp
                pasos+=2
                intercambios+=1
            pasos+=2
    info = [pasos, comparaciones, intercambios]
    return info
```

```python
n= 8
lista=[]
arreglopasos=[]
arreglocomparaciones=[]
arreglointercambios=[]
#todas las posibles permutaciones
for i in range(n):
    lista.append(i)
permutations=(list(itertools.permutations(lista)))
for sorting in permutations:
    arr=list(sorting)
    analisis=bubbleSort(arr)
    arreglopasos.append(analisis[0])
    arreglocomparaciones.append(analisis[1])
    arreglointercambios.append(analisis[2])


pl.hist(arreglointercambios, bins='auto')
pl.title("Histograma con todas las permutaciones de n intercambios")
pl.show()

pl.hist(arreglopasos, bins='auto')
pl.title("Histograma con todas las permutaciones de n pasos")
pl.show()

pl.hist(arreglocomparaciones, bins='auto')
pl.title("Histograma con todas las permutaciones de n comparaciones")
pl.show()
```
