```python
import numpy as np
import pylab as pl
import time
```

```python
n=100

listaprom = []

for i in range(2, int(n)+1):
    matrizA = np.full((i,i),1)
    matrizB = np.full((i,i),2)
    operaciones = 0
    for j in range(10):
        tempo0 = time.clock()
        matrizC = np.dot(matrizA, matrizB)
        tempo1 = time.clock()
        ops = (tempo1-tempo0)/(2*int(i)**3)
        operaciones+=ops

    listaprom.append(operaciones/10)
numeros = np.arange(1,int(n))
pl.ylabel("Tiempo")
pl.xlabel("N Matrices, N Tamano")
pl.plot(numeros, listaprom)
pl.show()
```
