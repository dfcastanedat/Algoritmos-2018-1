```python
import time
import numpy as np
import pylab as pl
```

```python
def fibM(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b


tiempos = []
tiempoinicial = time.clock()
for i in range (0, 3500, 1):
    tiempofinal = time.clock()
    tiempototal = tiempofinal - tiempoinicial
    print ("N=",i,"Fibb de N=",fibM(i),"Con tiempo de:",tiempototal)
    tiempos.append(tiempototal)
numeros = np.arange(0,3500,1)
pl.ylabel("Tiempo")
pl.xlabel("Fibb de N")
pl.plot(numeros,tiempos)
pl.show()
```
